```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('punkt')
all_complaints = ' '.join(df['Жалобы'])
tokens = word_tokenize(all_complaints)
fdist = FreqDist(tokens)
filtered_complaints = {word: count for word, count in fdist.items() if len(word) > 3 and word.lower() != 'Жалобы' and word.lower() != 'области' and word.lower() != 'осмотра' and word.lower() != 'Жалобы' and word.lower() != 'баллов' and word.lower() != 'момент' and word.lower() != 'предъявляет' and word.lower() != 'жалоб' and word.lower() != 'шкале'}
sorted_complaints = sorted(filtered_complaints.items(), key=lambda x: x[1], reverse=True)[:10]
for word, count in sorted_complaints:
    print(f"('{word}', {count}),")
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from natasha import MorphVocab
nltk.download('punkt')
morph_vocab = MorphVocab()
all_complaints = ' '.join(df['Жалобы'])
tokens = word_tokenize(all_complaints)

normalized_words = {}

for token in tokens:
    normal_form = morph_vocab(token)[0].normal_form
    normalized_words[token] = normal_form

filtered_complaints = {word: count for word, count in fdist.items() if len(word) > 3 and normalized_words.get(word.lower(), '') not in ['жалобы', 'области', 'осмотра', 'баллов', 'момент', 'предъявляет', 'жалоб', 'шкале']}
sorted_complaints = sorted(filtered_complaints.items(), key=lambda x: x[1], reverse=True)[:10]
for word, count in sorted_complaints:
    print(f"('{word}', {count}),")
```

```python
from natasha import Doc, Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger
top_10_complaints = ['боли', 'боль', 'дискомфорт', 'слабость', 'горле', 'носа', 'живота', 'кашель', 'заложенность']
segmenter = Segmenter()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
morph_vocab = MorphVocab()
phrases = {complaint: set() for complaint in top_10_complaints}
for complaint in top_10_complaints:
    for text in df['Жалобы']:
        doc = Doc(text)
        doc.segment(segmenter)
        doc.tag_morph(morph_tagger)
        if doc.spans:
            for span in doc.spans:
                if span.type == 'NOUN' and morph_vocab(span.text)[0].normal_form == morph_vocab(complaint)[0].normal_form:
                    phrases[complaint].add(span.text)
for complaint, extracted_phrases in phrases.items():
    print(f"{complaint}: {', '.join(extracted_phrases)}")
```
